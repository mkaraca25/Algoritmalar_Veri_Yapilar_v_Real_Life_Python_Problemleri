# Dynamic Array

```python
import ctypes #yeni array yaratmak iÃ§in kullanilir.

class DynamicArray(object):
    def __init__(self):
        self.n=0       #eleman sayisi
        self.capacity=1 #kapasite
        self.A=self.make_array(self.capacity)
    def __len__(self):
        return self.n
    def __getitem__(self,k):
        if not 0<= k < self.n:
            return IndexError("k is out of bounds !")
        return self.A[k]
    def append(self,eleman):
        if self.n==self.capacity:
            self.resize(2*self.capacity)
        self.A[self.n]=eleman
        self.n+=1
    def resize(self,new_cap):
        B=self.make_array(new_cap)
        
        for k in range(self.n):
            B[k]=self.A[k]
        self.A=B
        self.capacity=new_cap
    def make_array(self,new_cap):
        return (new_cap*ctypes.py_object)() #yeni array
```

```python
arr=DynamicArray()
arr.append(1)
print(arr[0])
arr.append(3)
print(arr[0],arr[1])
arr.append(5)
print(arr[0],arr[1],arr[2])
```

# wordSplit
*input=["deeplearning","d,dll,a,dee,base,lear,learning"]

 *output=["deep, learning"]

```python
def wordSplit(liste):
    word=list((liste[0]))
    d=liste[1].split(",")
    for i in range(1,len(word)):
        c=word[:]
        c.insert(i," ")
        x , y="".join(c).split()
        if x in d and y in d:
            return x + " , "+ y
    return "bulunamadi..."
```

```python
print(wordSplit(["deeplearning","d,dll,a,deep,dee,base,lear,learning"]))
```
