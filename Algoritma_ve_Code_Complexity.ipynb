```python
def square_sum(n):
    total=0
    for x in range(1,n+1):
        total+=x*x
    return total
print(square_sum(5))
```

```python
def square_sum1(n):

    return ((n*(n+1)*(2*n+1))/6)
square_sum1(5)
```

# Big-O Ornekleri

```python
def constant_big_O(liste):
    print(liste[0])
constant_big_O([1,2,3,4,5,6])
```

```python
def linear_big_O(liste):
    for deger in liste:
        print(deger)
linear_big_O([1,2,3,4,5,6])
```

```python
def cubic_big_O(liste):
    for item_1 in liste:
        for item_2 in liste:
            for item_3 in liste:
                print(item_1,item_2,item_3)
cubic_big_O([1,2,3])                
```

```python
def linear_big_O(liste):
    for deger in liste: #O(n)
        print(deger)
```

```python
def linear_big_O(liste):
    for deger in liste:
        print(deger)
    for deger in liste:   #O(2n)
        print(deger)    
```

```python
#O(1+n)
def example(liste):
    print(liste[0]) #O(1) constant
    for each in liste:
        print(each)   #O(n) linear
example([1,2,3,4,5])
```
