## Stacks - Queues - Deques

```python
class Stack:
    
    def __init__(self):
        self.items=[]
    def isEmpty(self):
        """
            boÅ olup olmadigini kontrol eder.
        """
        return self.items==[]
    def push(self,item):
        self.items.append(item) #item ekleme
    def pop(self):
        self.items.pop() #item cikarma
    def top(self):
        return self.items[len(self.items)-1] # son elemani gosterme
    
    def size(self):
        return len(self.items)
```

```python
stack=Stack()
print(stack.isEmpty())
stack.push("Erzurum")
print(stack.top())
stack.push("Ankara")
print(stack.top())
print(stack.size())
```

```python
class Queu:
    def __init__(self):
        self.items=[]
    def isEmpty(self):
        return self.items==[]
    def enqueue(self,item):
        self.items.insert(0,item)
    def dequeue(self):
        return self.items.pop()
    def size(self):
        return len(self.items)
```

```python
queue=Queu()
print(queue.isEmpty())
queue.enqueue("Erzurum")
queue.enqueue("Ankara")
queue.enqueue("Samsun")
print(queue.size())
queue.dequeue()
print(queue.size())
```

# deque

```python
class Deque:
    def __init__(self):
        self.items=[]
    def isEmpty(self):
        return self.items==[]
    def addFront(self,item):
        self.items.append(item)
    def addRear(self,item):
        self.items.insert(0,item)
    def removeFront(self):
        return self.items.pop()
    def removeRear(self):
        return self.items.pop(0)
    def size(self):
        return len(self.items)
```

```python
deque=Deque()
print(deque.isEmpty())
deque.addFront("ERZURUM")
deque.addRear("ANKARA")
print("size : ",deque.size())
deque.removeFront()
print("size : ",deque.size())
```

```python
#iki stack kullanarak queue yaratma
class Queue2Stack:
    def __init__(self):
        self.stack=[]
        self.stack1=[]
    def enqueue(self,item):
        self.stack.append(item)
    def dequeue(self):
        if not self.stack1:
            while len(self.stack)>0:
                self.stack1.append(self.stack.pop())
        return self.stack1.pop()
```

```python
queue=Queue2Stack()
queue.enqueue("1")
queue.enqueue("2")
queue.enqueue("3")
print(queue.dequeue())
print(queue.dequeue())
```
